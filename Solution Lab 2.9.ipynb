#### 1. How many distinct (different) actors' last names are there?

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
query = '''SELECT 
COUNT(DISTINCT last_name) AS distinct_surnames 
FROM actor;'''
data = pd.read_sql_query(query, engine)
data.head()
```

#### 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
query = '''SELECT *, MONTHNAME(rental_date) AS month, 
DAYNAME(rental_date) AS weekday,
CASE
WHEN DAYNAME(rental_date)
IN ('Saturday', 'Sunday')
THEN 'Weekend' 
ELSE 'Weekday'
END
AS day_type FROM rental;'''
data = pd.read_sql_query(query, engine)
display(data.head())
display(data.tail())
```

#### 3. Get all films with ARMAGEDDON in the title.

```python
query = '''SELECT * 
FROM film 
WHERE title LIKE '%%ARMAGEDDON%%';'''
data = pd.read_sql_query(query, engine)
data.head(10)
```

#### 4. Get 10 the longest films.

```python
query = '''SELECT title, length 
FROM film 
ORDER BY length 
DESC LIMIT 10;'''
data = pd.read_sql_query(query, engine)
data.head(10)
```

#### 5. How many films include Behind the Scenes content?

```python
query = '''SELECT COUNT(title) 
FROM film 
WHERE special_features 
LIKE '%%Behind the Scenes%%';'''
data = pd.read_sql_query(query, engine)
data.head()
```

#### 6. Which kind of movies (rating) have a mean duration of more than two hours?

```python
query = '''SELECT 
    rating, ROUND(AVG(length), 2) AS average_length
FROM
    film
GROUP BY rating
HAVING AVG(length) > 120;'''
data = pd.read_sql_query(query, engine)
data.head()
```

#### 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
query = '''SELECT title,length,
RANK () OVER (ORDER BY length DESC)
AS ranks 
FROM film
WHERE length !=0 AND length IS NOT NULL
ORDER BY length DESC;'''
data = pd.read_sql_query(query, engine)
data.head()
```
